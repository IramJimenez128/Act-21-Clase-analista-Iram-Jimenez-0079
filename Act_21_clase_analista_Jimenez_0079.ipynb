```python
class Analista:
    def __init__(self, afinidad_tecnica, comunicacion_efectiva, conocimiento_del_dominio):
        self.afinidad_tecnica = afinidad_tecnica
        self.comunicacion_efectiva = comunicacion_efectiva
        self.conocimiento_del_dominio = conocimiento_del_dominio
        self.activo = False

    def recopilacion(self):
        self.activo = True
        print(f"El analista con afinidad {self.afinidad_tecnica} ha comenzado la recopilación.")

    def analisis(self):
        print(f"Realizando análisis sobre: {self.conocimiento_del_dominio}.")

    def visualizacion(self):
        print(f"Generando reportes con nivel de comunicación: {self.comunicacion_efectiva}.")

# Instanciación
analista1 = Analista("Python", "Alta", "Finanzas")
analista2 = Analista("SQL", "Media", "Marketing")

# Llamadas a las funciones para cada objeto analista 1 y analista 2
print("-" * 50)
print("Analista 1: ")
analista1.recopilacion()
analista1.analisis()
analista1.visualizacion()

print("-" * 50)
print("Analista 2: ")
analista2.recopilacion()
analista2.analisis()
analista2.visualizacion()

print("-" * 50)
print("Programa terminado")
print("Creador de Software por Iram Jimenez NL = 28")
```
